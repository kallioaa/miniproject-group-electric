# Simple Regression Model (RandomForestTree)

```python
import sklearn
import numpy as np
import pandas as pd
```

```python
# importing dataframe
df = pd.read_csv("../data/data_2016-2021_hourly_sifted.csv", sep=";")

print(df.head(1))
```

### Split the data into training data and test data

```python
from sklearn.model_selection import train_test_split

df_copy = df.copy()

y = df_copy['PRICE (EUR/MWh)']
x = df_copy.drop(columns=['PRICE (EUR/MWh)', 'PRODUCTION (MWh)', 'CONSUMP (MWh)', "Year", "Hour", "Month", "Day"], axis=1)

x_train, x_test, y_train, y_test = train_test_split(x, y)

print("Training data: ")
print(" (x) " + str(x_train.shape))
print(" (y) " + str(y_train.shape))
print()
print("Test data: ")
print(" (x) " + str(x_test.shape))
print(" (y) " + str(y_test.shape))
```

### Create a model and train it

```python
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor

regr = RandomForestRegressor()
regr.fit(x_train, y_train)
```

### Evaluate the performance

```python
regr.score(x_test, y_test)
```
